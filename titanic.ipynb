# Identification of outliers and Removing them using IQR method

```python
#bocplot of dataset 
sns.boxplot(df)
```

```python
#Box plot for each column 
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.boxplot(data=df, x='Survived', palette = 'Blues',ax=axes[0, 0])
axes[0, 0].set_title('Box Plot for survived')

sns.boxplot(data=df, x='Pclass',palette ='husl', ax=axes[0, 1])
axes[0, 1].set_title('Box Plot for Pclass')

sns.boxplot(data=df, x='Sex',palette ='deep', ax=axes[1, 0])
axes[1, 0].set_title('Box Plot for Sex')

sns.boxplot(data=df, x='Age',palette ='viridis', ax=axes[1, 1])
axes[1, 1].set_title('Box Plot for Age')

plt.tight_layout()
plt.show()
```

```python
# identifying the outliers in the column Age
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3-Q1
threshold=1.5 
admit_outliers=df[(df['Age'] <Q1-threshold*IQR) | (df['Age'] > Q3+ threshold*IQR)]
admit_outliers
```

```python
# Drop the outliers
df=df.drop(admit_outliers.index)
sns.boxplot(x='Age',data=df)
```

```python
# identifying the outliers in the column SibSp
Q1 = df['SibSp'].quantile(0.25)
Q3 = df['SibSp'].quantile(0.75)
IQR = Q3-Q1
threshold=1.5 
admit_outliers=df[(df['SibSp'] <Q1-threshold*IQR) | (df['SibSp'] > Q3+ threshold*IQR)]
admit_outliers
```

```python
# Drop the outliers
df=df.drop(admit_outliers.index)
sns.boxplot(x='SibSp',data=df)
```

```python
# identifying the outliers in the column Parch
Q1 = df['Parch'].quantile(0.25)
Q3 = df['Parch'].quantile(0.75)
IQR = Q3-Q1
threshold=1.5 
admit_outliers=df[(df['Parch'] <Q1-threshold*IQR) | (df['Parch'] > Q3+ threshold*IQR)]
admit_outliers
```

```python
# Drop the outliers
df=df.drop(admit_outliers.index)
sns.boxplot(x='Parch',data=df)
```

# Splitting of x and y

```python
#spliting the data into x and y
x=df.iloc[:,1:6]
y=df.iloc[:,-0]
x
```

```python
#import train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=11)
x_train
```

```python
y_train
```

```python
#Transforming the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #removes the mean and scales each feature/variable to unit variance.
x_train = scaler.fit_transform(x_train) 
x_test = scaler.transform(x_test)
x_test
```

# Logistic Regression

```python
#import the logistic Regression library
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
#Assinging the parameter
from sklearn.model_selection import GridSearchCV
parameter ={'penalty':['l1','l2','elasticnet'],
             'C':[1,2,3,4,5,6,10,20,30,40,50],
              'max_iter':[100,200,300]}

classifier_regressor = GridSearchCV(classifier,
                                    param_grid = parameter,
                                    scoring='accuracy',
                                    cv = 5)

classifier_regressor.fit(x_train, y_train)
print(classifier_regressor.best_score_)
```

```python
#predicting x test
classifier_regressor.predict(x_test)
```

```python
#Predicting x_test
classifier_regressor.predict(x_test)
```

```python
# assigning the y_pred 
y_pred = classifier_regressor.predict(x_test)
y_pred
```

```python
#calculating the accuracy
from sklearn.metrics import accuracy_score,classification_report
score_lr = accuracy_score(y_pred,y_test)
print("Accuracy:",score_lr*100,"%")
```

```python
#Predicting the result
lr = classifier_regressor.predict([[3,0,40,1,1]]) 
lr
```
